## Assess

Please note that before I do any deep assessments of the data, I fill in missing data by column using 'bfill limit 1' and 'ffill limit 1' in pd.fillna() to get rid of as many nans as I can. I then reassess the remaining nans.

*(1) The remaining nans are a small percentage of each column, except for the tags, which are custom, and therefore cannot be addressed. The nulls will be left alone.  
*(2) Some of the sentences in the 'use' column are not capitalized.

#### Assessment 1

```python
# Fill in nans in each column by using the previous and then next available value, limit 1 in a row.
df.fillna(method = 'ffill', axis = 0, inplace = True, limit = 1)
df.fillna(method = 'bfill', axis = 0, inplace = True, limit = 1)
```
